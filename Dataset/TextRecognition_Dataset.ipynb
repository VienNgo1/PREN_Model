# Regular dataset

## SVT (Street View Text Dataset)

**Data type**: regular

**Number of source images**: 350 images

**Annotation type**: xml - Alex Sorokin's Annotation Toolkit

**Original Source**: http://www.iapr-tc11.org/mediawiki/index.php/The_Street_View_Text_Dataset

**Processed**: https://nhidang-text-recognition.s3.ap-southeast-1.amazonaws.com/SVT.zip

**After processing**:
  *   test: 648 images
  *   train: 258 images




## IC03 (ICDAR 2003 Robust Reading Competitions)

**Data type**: regular

**Number of source images**: 258 train + 251 test

**Annotation type**: xml

**Original Source**: http://iapr-tc11.org/mediawiki/index.php?title=ICDAR_2003_Robust_Reading_Competitions

**Processed**: https://nhidang-text-recognition.s3.ap-southeast-1.amazonaws.com/IC03.zip

# Irregular dataset

## IC15 (ICDAR 2015 Robust Reading Competitions)


**Data type**: irregular - text images were taken by Google Glasses without ensuring the image quality, most of the text is very small, blurred, and multi-oriented

**Number of source images**: 1000 train + 500 test

**Annotation type**: processed files avaiable online

**Original Source**: https://rrc.cvc.uab.es/?ch=4&com=downloads

**Processed**:
  *   https://mmocr.readthedocs.io/en/v0.3.0/datasets/recog.html
  *   https://nhidang-text-recognition.s3.ap-southeast-1.amazonaws.com/IC15.zip (already collected test/train labels)

**After processing**:
  *   test: 2077 images
  *   train: 4468 images

## SVTP (Street View Text Perspective )

**Data type**: irregular - most text instances are heavily distorted by the non-frontal view angle

**Number of source images**: 238 test

**Annotation type**: processed files avaiable online

**Original Source**: not available

**Processed**:
  *   https://github.com/Jyouhou/Case-Sensitive-Scene-Text-Recognition-Datasets
  *   https://nhidang-text-recognition.s3.ap-southeast-1.amazonaws.com/SVTP.zip (already combine all labels into one txt file)

**After processing**:
  *   test: 639 images

## CUTE (Curve Text - CUTE80)

**Data type**: irregular - focuses on curved text recognition. Most images in CUTE80 have a complex background, perspective distortion, and poor resolution

**Number of source images**: 80 test

**Annotation type**: processed files avaiable online

**Original Source**: http://cs-chan.com/downloads_CUTE80_dataset.html

**Processed**:
  *   https://github.com/Jyouhou/Case-Sensitive-Scene-Text-Recognition-Datasets
  *   https://nhidang-text-recognition.s3.ap-southeast-1.amazonaws.com/CUTE.zip (already combine all labels into one txt file)

**After processing**:
  *   test: 289 images
